```python
import matplotlib.pyplot as plt
```

```python
model_vals = {"x": [1.125, 1.35, 2.13, 2.27, 3.13, 3.30, 3.66],
              "y": [41.40, 52.33, 45.66, 47.67, 46.60, 43.0, 41.08]
              }
plt.plot(model_vals["x"], model_vals["y"], marker="o")
plt.xticks([1.0, 2.0, 3.0, 4.0],
           [r'$10^{2}$', r'$10^{3}$', r'$10^{4}$', r'$10^{5}$'],
           fontsize=15)
plt.yticks(fontsize=15)
plt.grid()
plt.xlabel("Parameters (Million)", fontsize=15)
plt.ylabel("Context Influence", fontsize=15)
plt.savefig("../plots/model_size.pdf", bbox_inches="tight")
```

```python
model_vals = {"x": [1.06, 1.25, 1.5, 2, 3, 4],
              "y": [25.67, 41.16, 44.50, 44.01, 44.62, 45.66]
              }
plt.plot(model_vals["x"], model_vals["y"], marker="o")
plt.xticks([1.0, 2.0, 3.0, 4.0],
           [0, 512, 1024, 2048],
           fontsize=15)
plt.yticks([24, 30, 36, 42, 48], fontsize=15)
plt.grid()
plt.xlabel("Context Size", fontsize=15)
plt.ylabel("Context Influence", fontsize=15)
plt.savefig("../plots/context_size.pdf", bbox_inches="tight")
```
